```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

sns.histplot(df['popularity'], ax=axes[0], color='steelblue')
axes[0].set_title('Popularity fördelning')

sns.histplot(df['vote_average'], ax=axes[1], color='coral')
axes[1].set_title('Betyg fördelning')

sns.histplot(df['vote_count'], ax=axes[2], color='seagreen')
axes[2].set_title('Antal röster fördelning')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='popularity', y='vote_average', 
                size='vote_count', sizes=(50, 500), alpha=0.7)


for _, row in df.iterrows():
    plt.annotate(row['title'], (row['popularity'], row['vote_average']),
                 fontsize=7, alpha=0.8, xytext=(5, 5), 
                 textcoords='offset points')

plt.title('Popularitet vs Betyg (storlek = antal röster)')
plt.xlabel('Popularity')
plt.ylabel('Betyg')
plt.tight_layout()
plt.show()
```

# Release år

```python
plt.figure(figsize=(10, 6))
df['release_year'] = pd.to_datetime(df['release_date']).dt.year
df['release_year'].value_counts().sort_index().plot(kind='bar', color='coral')
plt.title('Filmer per år')
plt.xlabel('År')
plt.ylabel('Antal filmer')
plt.tight_layout()
plt.show()
```

# Språkfördelning

# Top 10 betyg

```python
plt.figure(figsize=(10, 6))
top_rating = df.nlargest(10, 'vote_average')[['title', 'vote_average']]
sns.barplot(data=top_rating, x='vote_average', y='title', color='steelblue')
plt.title('Top 10 – Högst betyg')
plt.xlabel('Betyg')
plt.ylabel('')
plt.xlim(5, 9)
plt.tight_layout()
plt.show()
```

# Top 10 popularity

```python
plt.figure(figsize=(10, 6))
top_pop = df.nlargest(10, 'popularity')[['title', 'popularity']]
sns.barplot(data=top_pop, x='popularity', y='title', color='coral')
plt.title('Top 10 – Mest populära')
plt.xlabel('Popularity')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python

genre_map = {
    28: 'Action', 12: 'Äventyr', 14: 'Fantasy',
    878: 'Sci-Fi', 16: 'Animerat', 35: 'Komedi',
    18: 'Drama', 27: 'Skräck', 10751: 'Familj'
}

df['genres'] = df['genre_ids'].apply(lambda x: [genre_map.get(i, str(i)) for i in x])
all_genres = [g for sublist in df['genres'] for g in sublist]
pd.Series(all_genres).value_counts().plot(kind='bar', color='purple')
plt.title('Genrefördelning')
plt.tight_layout()
plt.show()
```

# vilken månad släpps mest fantasy
